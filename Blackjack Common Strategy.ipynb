```python
fileSoft = "./Blackjack_Common_Strategy/soft_totals_policy.csv"
def softTotalsPolicy(file):
    soft_totals_policy_dict = {}
    with open(file, newline='') as csvfile:
        soft_totals_policy_reader = csv.reader(csvfile, delimiter=',')
        line_num = 0
        for row in soft_totals_policy_reader:
            line_num += 1
            if line_num == 1:
                dealerCard_list = row[1:]
                for dealerCard in dealerCard_list:
                    soft_totals_policy_dict[dealerCard] = {} 
            else:
                player_first2cards = row[0]
                for dealerCard_index in range(len(dealerCard_list)):
                    player_action = row[dealerCard_index + 1]
                    dealerCard = dealerCard_list[dealerCard_index]
                    soft_totals_policy_dict[dealerCard][player_first2cards] = player_action
            
    
    return soft_totals_policy_dict
```

```python
# Return the common player strategy when the initial cards contain Ace and when 
# we have not hit before. The common strategy is based on the soft total table.
# Return 1 if hitting, 0 if standing and 2 if doubling when doubling is allowed.
def softStrategy(dealerCard1, playerHistory, doublingAllowed, softTotalsPolicyDict):
    
    playerHistoryDict = copy.deepcopy(playerHistory)
    dealerCard1Dict = [dealerCard1]
    playerHistoryDict.sort()
    dealerCard1Dict.sort()
    
    for i in range(len(playerHistoryDict)):
        if playerHistoryDict[i] == 1:
            playerHistoryDict[i] = 'A'
        elif playerHistoryDict[i] == 11 or playerHistoryDict[i] == 12 or playerHistoryDict[i] == 13:
            playerHistoryDict[i] = 10
    
    for i in range(len(dealerCard1Dict)):
        if dealerCard1Dict[i] == 1:
            dealerCard1Dict[i] = 'A'
        elif dealerCard1Dict[i] == 11 or dealerCard1Dict[i] == 12 or dealerCard1Dict[i] == 13:
            dealerCard1Dict[i] = 10
            
    # turn dealerCard and playerHistory from list type to string so that 
    # we can extract policy from the dictionary.
    playerHistoryString = [str(x) for x in playerHistoryDict]
    playerHistoryString = ','.join(playerHistoryString)
    
    dealerCard1String = [str(x) for x in dealerCard1Dict]
    dealerCard1String = ','.join(dealerCard1String)
    
    result = softTotalsPolicyDict[dealerCard1String]
    #print(dealerCard1String, result, playerHistoryString)
    strategyString = result[playerHistoryString]
    
    if strategyString == 'S':
        strategy = 0
    elif strategyString == 'H':
        strategy = 1
    elif strategyString == 'D':
        if doublingAllowed and len(playerHistory):
            strategy = 2
        else:
            strategy = 1
    elif strategyString == 'Ds':
        if doublingAllowed and len(playerHistory):
            strategy = 2
        else:
            strategy = 0
            
    return strategy
```

```python
# Return the common player strategy when the initial cards does not contain Ace 
# or when we have hit before. The common strategy is based on the hard total table.
# Return 1 if hitting, 0 if standing and 2 if doubling when doubling is allowed.
def hardStrategy(dealerCard1, playerHistory, doublingAllowed, hardTotalsPolicyDict):
    
    strategy = 0
    
    playerSum = valueSum(playerHistory, 0)

    if isinstance(playerSum, int) == False:
        playerSumString = str(playerSum[1])
    else:
        playerSumString = str(playerSum)
    
    dealerCard1Dict = [dealerCard1]
    for i in range(len(dealerCard1Dict)):
        if dealerCard1Dict[i] == 1:
            dealerCard1Dict[i] = 'A'
        elif dealerCard1Dict[i] == 11 or dealerCard1Dict[i] == 12 or dealerCard1Dict[i] == 13:
            dealerCard1Dict[i] = 10
    dealerCard1String = [str(x) for x in dealerCard1Dict]
    dealerCard1String = ','.join(dealerCard1String)
    
    strategyString = hardTotalsPolicyDict[dealerCard1String][playerSumString]
    
    if strategyString == 'S':
        strategy = 0
    elif strategyString == 'H':
        strategy = 1
    elif strategyString == 'D':
        if doublingAllowed and len(playerHistory):
            strategy = 2
        else:
            strategy = 1
    elif strategyString == 'Ds':
        if doublingAllowed and len(playerHistory):
            strategy = 2
        else:
            strategy = 0
            
    return strategy
    
```

```python
# doublingAllowed is a boolean value. It is True when doubling is allowed.
# Return the common player strategy used. This does not include splitting.
# Return 1 if hitting, 0 if standing and 2 if doubling when doubling is allowed.
def commonStrategy(dealerCard1, playerHistory, doublingAllowed, softTotalsPolicyDict, hardTotalsPolicyDict):
    if len(playerHistory) == 2 and (1 in playerHistory):
        return softStrategy(dealerCard1, playerHistory, doublingAllowed, softTotalsPolicyDict)
    else:
        return hardStrategy(dealerCard1, playerHistory, doublingAllowed, hardTotalsPolicyDict)
```

```python
# Return the percentage of winning games and the money left if
# doubling is allowed.
def commonPolicySimulation(n, dealerCard1, dealerCard2, playerCard1, 
                 playerCard2, doublingAllowed, softTotalsPolicyDict, hardTotalsPolicyDict):
    
    cards = [1] * 52
    cardsCount = [4] * 13
    
    playerHistory = [playerCard1, playerCard2]
    dealerHistory = [dealerCard1, dealerCard2]
    
    for i in range(0, len(playerHistory)):
        hit(playerHistory[i], cards, cardsCount)
    for i in range(0, len(dealerHistory)):
        hitResult = hit(dealerHistory[i], cards, cardsCount)
    
    # the player get 0 money at the start of the simulation
    currentMoney = 0
    numOfWin = 0
    
    for i in tqdm(range(0, n)):
        
        cards1 = copy.deepcopy(hitResult[1])
        cardsCount1 = copy.deepcopy(hitResult[2])
        playerHistory = [playerCard1, playerCard2]
        
        playerSum = valueSum(playerHistory, 0)
        if isinstance(playerSum, int) == False:
            playerSum = playerSum[1]
        
        strategy = commonStrategy(dealerCard1, playerHistory, doublingAllowed, 
                                  softTotalsPolicyDict, hardTotalsPolicyDict)
        
        while strategy == 1:
            value = randomValue(cards1) 
            hit(value, cards1, cardsCount1)
            playerHistory.append(value)
            
            playerSum = valueSum(playerHistory, 0)
            if isinstance(playerSum, int) == False:
                playerSum = playerSum[1]
            if playerSum > 21:
                break
            
            strategy = commonStrategy(dealerCard1, playerHistory, doublingAllowed, 
                                      softTotalsPolicyDict, hardTotalsPolicyDict)
        
        if playerSum > 21:
            currentMoney -= 1
            continue
        
        # for doubling game, the player wins or loses 2 units of money.
        if strategy == 2:
            value = randomValue(cards1) 
            hit(value, cards1, cardsCount1)
            playerHistory.append(value)
            
            playerSum = valueSum(playerHistory, 0)
            if isinstance(playerSum, int) == False:
                playerSum = playerSum[1]
            
            if playerSum > 21:
                currentMoney -= 2
                continue
        
        if strategy == 0:
            playerSum = valueSum(playerHistory, 0)
            if isinstance(playerSum, int) == False:
                playerSum = playerSum[1]
        
        dealerHistory = [dealerCard1, dealerCard2]
        dealerSum = valueSum(dealerHistory, 1)

        while dealerSum < 17:
            value = randomValue(cards1) 
            hit(value, cards1, cardsCount1)
            dealerHistory.append(value)
            dealerSum = valueSum(dealerHistory, 1)
        
        if dealerSum > 21 or playerSum > dealerSum:
            numOfWin += 1
            if strategy == 2:
                currentMoney += 2
            else:
                currentMoney += 1
    return numOfWin / n, currentMoney
        
```

```python
n = 100000

dealerCard1 = 1
dealerCard2 = 6

playerCard1 = 5
playerCard2 = 9

doublingAllowed = False

file = "./Blackjack_Common_Strategy/hard_totals_policy.csv"
fileSoft = "./Blackjack_Common_Strategy/soft_totals_policy.csv"

softTotalsPolicyDict = softTotalsPolicy(fileSoft)
hardTotalsPolicyDict = hardTotalsPolicy(file)

simulation = commonPolicySimulation(n, dealerCard1, dealerCard2, playerCard1, 
                 playerCard2, doublingAllowed, softTotalsPolicyDict, hardTotalsPolicyDict)

print(simulation[0], simulation[1])
```
